```python
#from https://chatgpt.com/c/68e517de-6ad0-8332-aa27-fa24fa918b8a

import json
import matplotlib.pyplot as plt
import numpy as np

# Load and clean JSON
with open("simpler.json", "r") as f:
    json_str = ""
    for line in f:
        if "//" in line:
            line = line[: line.index("//")]
        json_str += line
    vqa_dict = json.loads(json_str)

# Count how many questions are in each category
category_counts = {cat: len(vqa_dict[cat]) for cat in vqa_dict}
labels = list(category_counts.keys())
sizes = list(category_counts.values())

# Setup colors
colors = plt.cm.plasma(np.linspace(0, 1, len(labels)))

fig, ax = plt.subplots(figsize=(9, 6), subplot_kw=dict(aspect="equal"))

# Donut-style chart (but center will now match background)
wedges, _ = ax.pie(
    sizes,
    startangle=140,
    colors=colors,
    wedgeprops=dict(width=0.4, edgecolor="white", linewidth=1.5)
)

# Add a fake 3D shadow under the pie
for w in wedges:
    w.set_zorder(2)
shadow = plt.Circle((0, 0), 0.7, color="black", alpha=0.2, zorder=1)
ax.add_artist(shadow)

# Add labels (category + count + percentage)
bbox_props = dict(boxstyle="round,pad=0.3", fc="black", alpha=0.6)
kw = dict(arrowprops=dict(arrowstyle="-", color="white", lw=0.8),
          bbox=bbox_props, zorder=3, va="center", color="white")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = f"angle,angleA=0,angleB={ang}"
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(f"{labels[i]}\n{sizes[i]} ({sizes[i]/sum(sizes)*100:.1f}%)",
                xy=(x, y), xytext=(1.3*np.sign(x), 1.3*y),
                horizontalalignment=horizontalalignment, **kw)

# Make background transparent and clean
fig.patch.set_alpha(0.0)       # transparent figure background
ax.set_facecolor("none")       # transparent axes
plt.axis('equal')
plt.title("DL3DV VQA  Category Distribution (3D Donut Edition)",
          color="white", fontsize=14, pad=60)

# Save with transparent background
plt.savefig("dl3dv_vqa_3d_donut.png", transparent=True, dpi=300, bbox_inches='tight')
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# ---------- Load and clean JSON ----------
with open("simpler.json", "r") as f:
    json_str = ""
    for line in f:
        if "//" in line:
            line = line[: line.index("//")]
        json_str += line
    vqa_dict = json.loads(json_str)

# ---------- Collect distributions ----------
# 1. Categories
category_counts = {cat: len(vqa_dict[cat]) for cat in vqa_dict}

# 2. Question types
types = []
measurements = []
for category in vqa_dict:
    for question in vqa_dict[category]:
        q = vqa_dict[category][question]
        types.append(q.get("question_type", "Unknown"))
        measurements.append(q.get("measurement", "Unknown"))

type_counts = Counter(types)
measurement_counts = Counter(measurements)

# ---------- Prepare data ----------
datasets = [
    ("Category", category_counts, plt.cm.plasma),
    ("Question Type", type_counts, plt.cm.viridis),
    ("Measurement", measurement_counts, plt.cm.cool),
]

# ---------- Plot setup ----------
fig, axes = plt.subplots(1, 3, figsize=(18, 6), subplot_kw=dict(aspect="equal"))

for ax, (title, data, cmap) in zip(axes, datasets):
    labels = list(data.keys())
    sizes = list(data.values())
    colors = cmap(np.linspace(0, 1, len(labels)))

    wedges, _ = ax.pie(
        sizes,
        startangle=140,
        colors=colors,
        wedgeprops=dict(width=0.4, edgecolor="white", linewidth=1.5)
    )

    # Fake shadow for 3D illusion
    for w in wedges:
        w.set_zorder(2)
    shadow = plt.Circle((0, 0), 0.7, color="black", alpha=0.2, zorder=1)
    ax.add_artist(shadow)

    # Add labels
    bbox_props = dict(boxstyle="round,pad=0.3", fc="black", alpha=0.6)
    kw = dict(arrowprops=dict(arrowstyle="-", color="white", lw=0.8),
              bbox=bbox_props, zorder=3, va="center", color="white")

    total = sum(sizes)
    for i, p in enumerate(wedges):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
        ha = {-1: "right", 1: "left"}[int(np.sign(x))]
        connectionstyle = f"angle,angleA=0,angleB={ang}"
        kw["arrowprops"].update({"connectionstyle": connectionstyle})
        ax.annotate(f"{labels[i]}\n{sizes[i]} ({sizes[i]/total*100:.1f}%)",
                    xy=(x, y), xytext=(1.2*np.sign(x), 1.2*y),
                    horizontalalignment=ha, **kw)

    ax.set_title(title, color="white", fontsize=13, pad=12)
    ax.set_facecolor("none")

# ---------- Styling ----------
fig.patch.set_alpha(0.0)  # transparent background
plt.subplots_adjust(wspace=0.4)
plt.suptitle("DL3DV VQA  Distribution Overview (3D Donut Style)",
             color="white", fontsize=16, weight="bold", y=1.02)
plt.savefig("dl3dv_vqa_triptych.png", transparent=True, dpi=300, bbox_inches='tight')
plt.show()
```
